```python
#2.Study of python basic libraries
import math

num = 9
angle_degrees = 45

print("Square Root:", math.sqrt(num))
print("Factorial:", math.factorial(5))
print("Logarithm base 10:", math.log10(100))
print("Sine of 45 degrees:", math.sin(math.radians(angle_degrees)))
print("Value of Pi:", math.pi)
```

```python
import math

# Example numbers
num1 = 4.3
num2 = 9.7
num3 = -2.8
num4 = 5.0

# Applying ceil function
print("Ceil of", num1, "is:", math.ceil(num1))  # Output: 5
print("Ceil of", num2, "is:", math.ceil(num2))  # Output: 10
print("Ceil of", num3, "is:", math.ceil(num3))  # Output: -2
print("Ceil of", num4, "is:", math.ceil(num4))  # Output: 5 (unchanged)
```

```python
#numpy basics
import numpy as np

a = np.array([1, 2, 3])   # Create a rank 1 array
print(type(a))
print(a.shape)            # Prints "(3,)"
print(a[0], a[1], a[2])   # Prints "1 2 3"
a[0] = 5                  # Change an element of the array
print(a)                  # Prints "[5, 2, 3]"
b = np.array([[1,2,3],[4,5,6]])    # Create a rank 2 array
print(b.shape)                     # Prints "(2, 3)"
print(b[0, 0], b[0, 1], b[1, 0])   # Prints "1 2 4"

```
